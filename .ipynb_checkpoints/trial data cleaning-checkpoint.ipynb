```python
import pandas as pd
import re
```

```python
data = pd.read_csv('apple_smartprix_data.csv')
data
```

```python
data.info() 
```

```python
#Giving columns appropriate names
data = data.rename(columns={'Price': 'Price_INR', 'Spec1': 'network_type', 'Spec2': 'Processor', 'Spec3': 'RAM', 'Spec4': 'battery_capicity_MAH', 'Spec5': 'display_size_INCH'})
```

```python
data = data.dropna()
```

```python
#changing data type of Price feature
price = data['Price_INR']
price1=[]
for p in price:
    clean_price=''
    for i in p :
        if i.isnumeric():
            clean_price+=i 
    price1.append(int(clean_price))
data['Price_INR'] = price1
```

```python
#finding netwirk type of mobile
ntype = data['network_type']
cleaned_ntype=[]
for t in ntype:
    if '6G' in t:
        cleaned_ntype.append('6G')
    elif '5G' in t:
        cleaned_ntype.append('5G')
    else:
        cleaned_ntype.append('4G')
```

```python
data['network_type'] =  cleaned_ntype
```

```python
#Extract processor name
data['Processor'] = data['Processor'].str.split(',').str[0]
```

```python
#extract ram and storage
ram_storage = data['RAM']
ram = data['RAM'].str.split(',').str[0]
ram
```

```python
data['RAM_size'] = data['RAM'].str.extract(r'(\d+\s*GB)\s*RAM')
data['RAM_size']
```

```python
#3 GB RAM, 64 GB inbuilt
data['storage_capacity'] = data['RAM'].str.extract(r',\s*(.*?)\s*inbuilt')
for i in data['storage_capacity']:
    print(i)
```

```python
#extract display size
data['display_size_INCH'] = data['display_size_INCH'].str.extract(r'(.*?)\s*inches')
```

```python
data['display_size_INCH'] = data['display_size_INCH'].astype(float)
```

```python
#extract battery capacity
data['battery_capicity_MAH'] = data['battery_capicity_MAH'].str.extract(r'(.*?)\s*mAh')
```
